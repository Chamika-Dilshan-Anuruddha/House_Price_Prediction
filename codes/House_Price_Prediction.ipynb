```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from  sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_data = sklearn.datasets.load_boston()
```

## Load the data

```python
print(house_price_data)
```

```python
# load the dataset to pandas data frame

house_price_dataframe = pd.DataFrame(house_price_data.data, columns = house_price_data.feature_names )
```

```python
house_price_dataframe.head()
```

```python
# add the class colum to the data frame

house_price_dataframe['price'] = house_price_data.target
```

```python
house_price_dataframe.head()
```

## Understhand the dataset 

```python
house_price_dataframe.shape
```

```python
# check missing values

house_price_dataframe.isnull().sum()
```

```python
# statistical measures data 

house_price_dataframe.describe()
```

```python
# corelation between various features

correlation  = house_price_dataframe.corr()

# creat heat map of correlation

plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

## Split data 

```python
X =house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
# train and test data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model (XGBoost Regressor)

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

## Accuracy

```python
# evaluation

# predict on training data


training_data_prediction = model.predict(X_train)
```

```python
# R squared error

score_1 = metrics.r2_score(Y_train, training_data_prediction)

# mean abs error

score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error train data: ", score_1)
print("mean abs error train data: ", score_2)
```

```python
# predict on testing data

test_data_prediction = model.predict(X_test)


# test data accuracy

# R squared error

score_1 = metrics.r2_score(Y_test, test_data_prediction)

# mean abs error

score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error test data: ", score_1)
print("mean abs error test data: ", score_2)
```

```python
# scater plot

plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicting price")
plt.title("Actual price vs Predicted price")
plt.show()
```
